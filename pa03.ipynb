# Birthday paradox simulator

In this programming assignment, you will learn how to plot curves and run random experiments to analyze the probability of events of different sizes.

You will write three functions in Python:
  - **birthday_pair(n,t)**: after t trials, estimate the probability that, in a set of n randomly chosen people, at least two will share a birthday. 
  - **birthday_different(n,t)**: after t trials, it gives the probability that, in a set of n randomly chosen people, none of those people shares a birthday. Even though this function complements birthday_pair(n,t), don't call the previous one. Implement both procedures independently.
  - **birthday_buddy(n,t)**: after t trials, estimate the probability that, in a set of n randomly chosen people, at least one person shares a birthday with you. 

Note:
 * For simplicity, you can ignore February 29th.
 * You can reuse the functions available on https://docs.python.org/3/library/random.html
 

```python
# Run this cell to load the libraries that you will need
import matplotlib.pyplot as plt
import random
```

```python
def birthday_pair(n,t):
  count = 0
  for i in range(t):
    birthdays = []
    for i in range(n):
      birthdays.append(random.randint(1,365))
    if n != len(set(birthdays)):
      count += 1
  return count/t
```

```python
def birthday_different(n,t):
  count = 0
  for i in range(t):
    birthdays = []
    for i in range(n):
      birthdays.append(random.randint(1,365))
    if n == len(set(birthdays)):
      count += 1
  return count/t
```

```python
def birthday_buddy(n,t):
  count = 0
  for i in range(t):
    birthdays = []
    for i in range(n):
      birthdays.append(random.randint(1,365))
    if 344 in birthdays:
      count += 1
  return count/t
```

```python
import matplotlib.pyplot as plt
import random

# Run each random experiment multiple times. 
# Adjust the value of trials for a more smooth graph or faster compilation
trials = 100

# Adjusting the size for a better visualization
plt.figure(figsize=(12,9))

# naming the x axis
plt.xlabel('Number of people')
x = [x for x in range(366)]

# naming the y axis
plt.ylabel('Probability')

# Collect the probability of birthday twins
y1 = [birthday_pair(n,trials) for n in range(366)]

# Collect the probability of no matching among n people
y2 = [birthday_different(n,trials) for n in range(366)]

# Collect the probability of at least one person share a birthday with me
y3 = [birthday_buddy(n,trials) for n in range(366)]

# giving a title to my graph
plt.title('Birthday paradox!')

# plotting the birthday twin points 
plt.plot(x, y1, label = "Probability of birthday twins")
  
# plotting the no matching birthday points 
plt.plot(x, y2, label = "Probability of no matching pair")

# plotting the birthday buddy points 
plt.plot(x, y3, label = "Probability of birthday buddy")

# show a legend on the plot
plt.legend(loc='upper right')

# function to show the plot
plt.show()
```
